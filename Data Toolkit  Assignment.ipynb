 # Question 1 
 
  **What is the difference between multithreading and multiprocessing?**

  
**Answer** 

Multithreading and multiprocessing are two important techniques used in operating systems and programming to improve the performance of applications by executing tasks concurrently.

**Multithreading** refers to the execution of multiple threads within a single process. All threads share the same memory space and resources of the process, which makes communication between threads fast and efficient. Multithreading is lightweight and has low overhead because thread creation and context switching are faster. It is mainly suitable for I/O-bound tasks such as file handling, web requests, database operations, and user interface responsiveness. However, in Python, multithreading is limited by the Global Interpreter Lock (GIL), which prevents true parallel execution of CPU-bound tasks. Also, since threads share memory, a failure in one thread can affect the entire process.

**Multiprocessing** involves the execution of multiple independent processes, each having its own memory space and resources. Because processes are isolated from each other, multiprocessing provides better stability and fault tolerance. It allows true parallelism by utilizing multiple CPU cores, making it ideal for CPU-bound tasks like heavy computations, data analysis, image processing, and machine learning. However, multiprocessing has higher overhead compared to multithreading, as creating processes and inter-process communication requires more system resources and is relatively slower.



### Difference Between Multithreading and Multiprocessing

| Feature | Multithreading | Multiprocessing |
|------|---------------|----------------|
| **Basic idea** | Multiple threads run within one process | Multiple processes run independently |
| **Memory** | Threads share the same memory space | Each process has its own memory |
| **Speed** | Faster creation and context switching | Slower than threads |
| **Communication** | Easy (shared variables) | Harder (IPC, queues, pipes) |
| **Isolation** | Low (one thread crash can affect others) | High (one process crash doesn’t affect others) |
| **CPU utilization** | Limited by GIL in Python | Can use multiple CPU cores fully |
| **Best for** | I/O-bound tasks (file, network, API calls) | CPU-bound tasks (heavy computation) |
| **Overhead** | Low | High |


**In conclusion,** multithreading is best suited for I/O-bound tasks where fast communication and low overhead are required, whereas multiprocessing is preferred for CPU-bound tasks where true parallelism, better performance, and reliability are important.


# Question 2: 
**What are the challenges associated with memory management in Python?**


**Answer:**  

Memory management in Python is automatic and handled by the Python interpreter, mainly through reference counting and garbage collection. While this simplifies development, it also introduces several challenges.

One major challenge is **high memory overhead**. Python objects store additional metadata such as reference counts and type information, which increases memory usage compared to lower-level languages like C or C++.

Another challenge is **garbage collection complexity**. Python primarily uses reference counting, but it cannot handle circular references on its own. Although the cyclic garbage collector addresses this, it may not immediately free memory, leading to delayed memory release.

**Memory fragmentation** is also a concern. Python’s memory allocator can fragment memory over time, especially in long-running applications, which can reduce performance and increase memory consumption.

Python programmers also face issues with **memory leaks**, particularly when global variables, caches, or references to unused objects are not released properly. External libraries written in C may also cause leaks that Python’s garbage collector cannot detect.

**Inefficient handling of large data structures** is another challenge. Objects like lists, dictionaries, and DataFrames can consume large amounts of memory if not managed carefully.

Additionally, **limited control over memory allocation** makes it difficult for developers to optimize memory usage at a low level, as Python abstracts most memory operations.

In conclusion, while Python’s automatic memory management improves developer productivity, challenges such as memory overhead, fragmentation, garbage collection delays, and memory leaks must be carefully handled, especially in data-intensive and long-running applications.


# Question 3:

**Write a Python program that logs an error message to a log file when a division by zero exception occurs.**


**Answer**

```python
# Python program to log an error message when division by zero occurs

import logging

# Configure logging
logging.basicConfig(
    filename="error.log",
    level=logging.ERROR,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

try:
    a = int(input("Enter numerator: "))
    b = int(input("Enter denominator: "))
    result = a / b
    print("Result:", result)

except ZeroDivisionError as e:
    logging.error("Division by zero error occurred", exc_info=True)
    print("Error: Division by zero is not allowed.")
```

# Question 4:

**Write a Python program that reads from one file and writes its content to another file.**


**Answer**

```python
# Python program that creates a source file (if not present),
# reads from it, and writes its content to another file

# Create and write content to source file
with open("source.txt", "w") as source_file:
    source_file.write("This is sample content written to the source file.")

# Read from source file
with open("source.txt", "r") as source_file:
    content = source_file.read()

# Write to destination file
with open("destination.txt", "w") as destination_file:
    destination_file.write(content)

print("Source file created and content copied to destination file successfully.")
```

# Question 5: 
**Write a program that handles both IndexError and KeyError using a try-except block.**


**Answer:**


```python
# Python program to handle IndexError and KeyError using try-except block

data_list = [10, 20, 30]
data_dict = {"a": 1, "b": 2}

try:
    # This may raise IndexError
    print(data_list[5])
    
    # This may raise KeyError
    print(data_dict["c"])

except IndexError:
    print("IndexError occurred: List index is out of range.")

except KeyError:
    print("KeyError occurred: Key not found in dictionary.")
```

# Question 6:
 **What are the differences between NumPy arrays and Python lists?**

**Answer:**  

### Differences Between NumPy Arrays and Python Lists

NumPy arrays and Python lists are both used to store collections of data, but they differ significantly in terms of performance, functionality, and usage.

**Python lists** are built-in data structures that can store elements of different data types. They are dynamic in size and provide great flexibility. However, lists require more memory and are slower for numerical computations because operations are performed element by element.

**NumPy arrays**, on the other hand, store elements of the same data type, which allows them to be more memory-efficient. They support fast, vectorized operations that are executed at a lower level, making numerical computations much faster. NumPy arrays also provide a wide range of mathematical and scientific functions.



### Comparison Between NumPy Arrays and Python Lists

| Feature | Python Lists | NumPy Arrays |
|------|-------------|--------------|
| **Data type** | Can store different data types | Stores elements of the same data type |
| **Memory usage** | Uses more memory | More memory-efficient |
| **Performance** | Slower for numerical operations | Faster due to vectorized operations |
| **Operations** | Element-wise using loops | Supports vectorized operations |
| **Functionality** | Limited mathematical functions | Rich set of mathematical and scientific functions |
| **Dimensions** | Mostly one-dimensional (nested lists for multi-dimension) | Supports multi-dimensional arrays |
| **Broadcasting** | Not supported | Supported |
| **Use case** | General-purpose programming | Numerical and scientific computing |
| **Library** | Built-in Python feature | Requires NumPy library |
| **Speed optimization** | No low-level optimization | Optimized in C for high performance |



**In addition, NumPy arrays** support broadcasting, multidimensional data handling, and efficient slicing, which are not as powerful in Python lists.

**In summary, Python list**s are suitable for general-purpose programming and heterogeneous data, **while NumPy arrays** are preferred for numerical analysis, scientific computing, and large datasets due to their performance and efficiency.


# Question 7:

**Explain the difference between apply() and map() in Pandas.**


**Answer:**  Difference Between apply() and map() in Pandas

In Pandas, **apply()** and **map()** are used to apply functions to data, but they differ in scope, flexibility, and usage.

**map()** is used only with **Series** objects. It applies a function element-wise to each value in a Series. It is mainly used for simple transformations, value replacement, or mapping values using a dictionary or another Series. The output of `map()` is always a Series.

**apply()** can be used with both **Series** and **DataFrame** objects. When used on a Series, it behaves similarly to `map()` but allows more complex functions. When used on a DataFrame, `apply()` can operate along rows or columns using the `axis` parameter, making it more powerful and flexible. However, `apply()` is generally slower than `map()` because it works at a higher abstraction level.



### Comparison Between apply() and map() in Pandas

| Feature | apply() | map() |
|------|--------|------|
| **Applicable to** | Series and DataFrame | Series only |
| **Operation level** | Row-wise or column-wise operations | Element-wise operations |
| **Function complexity** | Can handle complex functions | Best for simple functions |
| **Axis support** | Supports `axis` parameter | No axis parameter |
| **Return type** | Series or DataFrame | Always returns a Series |
| **Performance** | Slower compared to map() | Faster for simple transformations |
| **Use case** | Complex transformations and aggregations | Value mapping and simple replacements |
| **Flexibility** | More flexible | Less flexible |


In summary, `map()` is simpler and faster for element-wise operations on a single Series, while `apply()` is more flexible and suitable for row-wise or column-wise operations on DataFrames.

**Create a histogram using Seaborn to visualize a distribution.**


**Answer:**


```python


import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Sample data
data = np.random.randn(1000)

# Create histogram
sns.histplot(data, bins=30, kde=True)

# Add labels and title
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram of Data Distribution")

# Show plot
plt.show()
```

# Question 10:
**Calculate the correlation matrix using Seaborn and visualize it with a heatmap**

**Answer:**

```python
# Calculate correlation matrix and visualize it using Seaborn heatmap

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create sample DataFrame
data = {
    "Age": [22, 25, 30, 35, 40],
    "Salary": [25000, 30000, 45000, 60000, 80000],
    "Experience": [1, 3, 5, 8, 12]
}

df = pd.DataFrame(data)

# Calculate correlation matrix
corr_matrix = df.corr()

# Create heatmap
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")

# Add title
plt.title("Correlation Matrix Heatmap")

# Show plot
plt.show()
```
